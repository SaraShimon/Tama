```python
import pandas as pd
import seaborn as snc
```

```python
tama_df = pd.read_csv('engineered_tama_data.csv')
```

```python
corr = tama_df.corr()
snc.heatmap(corr, annot=True)
```

```python

from sklearn.preprocessing import MinMaxScaler 
import numpy as np 
  
# copy the data 
norm_tama_df = tama_df.copy() 
  
# apply normalization techniques 
columns = ['longitude', 'latitude']
norm_tama_df[columns] = MinMaxScaler().fit_transform(np.array(norm_tama_df[columns]).reshape(-1,2)) 
  
norm_tama_df
```

```python
# years normalization
norm_tama_df['year'] = norm_tama_df['year'] - 2005
norm_tama_df
```

```python
corr = norm_tama_df.corr()
snc.heatmap(corr, annot=True)
```

```python
from sklearn.preprocessing import MinMaxScaler 
import numpy as np 
  
norm_all_tama_df = tama_df.copy()
norm_all_tama_df.drop(columns=['hebrew_city_name'], inplace=True)
norm_all_tama_df = MinMaxScaler().fit_transform(np.array(norm_all_tama_df).reshape(-1,len(norm_all_tama_df.columns))) 
norm_all_tama_df =pd.DataFrame(norm_all_tama_df, columns=['city_code','year','total_open_requests', 'total_requests', 'total_approved_requests','longitude', 'latitude'])
norm_all_tama_df
```

```python
corr = norm_all_tama_df.corr()
snc.heatmap(corr, annot=True)
```
